```python
'''
PSNR heatmap for different layers and neurons for relu network
'''
import scienceplots
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'layers': [2, 3, 4, 5, 6],
    32: [25.7, 25.55, 26.6, 26.46, 26.06],
    64: [25.4, 26.16, 26.72, 24.34, 24.91],
    128: [26.82, 27.23, 25.83, 26.04, 25.68],
    256: [26.68, 26.82, 24.35, 24.31, 25.95]
}

df = pd.DataFrame(data)

with plt.style.context('seaborn-paper'):
    plt.figure(figsize=(12, 10))
    plt.rcParams.update({"text.usetex": True,
                         "font.family": "Helvetica",})
    sns.heatmap(df.set_index('layers'), annot=True, cmap='hot', fmt='.2f', annot_kws={"size": 15})
    plt.xlabel('Neurons', fontdict={'fontsize': 17})
    plt.ylabel('Layers', fontdict={'fontsize': 17})
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title('PSNR heatmap by params setup', fontdict={'fontsize': 20})
    plt.facecolor = 'white'
    # plt.savefig('relu_psnr_heatmap.pdf')
    plt.show()
```
